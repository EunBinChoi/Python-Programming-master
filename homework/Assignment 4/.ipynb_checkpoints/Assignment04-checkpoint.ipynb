> 본 문제를 풀기 위해 s.txt 파일 내용을 사전에 확인힙니다.

1) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.

> cat dog <br>
> dog pig <br>
> ham bird <br>
> pig ham <br>

````python
with open('s.txt', 'r') as f:
    text = f.readlines()
    text.sort()
    
with open('s1.txt', 'w') as f:
    f.writelines(text)

# s1.txt 파일에 내용이 잘 들어갔는 지 확인
with open('s1.txt', 'r') as f:
    print(f.read()) 
````

> 1. 먼저 s.txt 파일을 읽기모드로 열어 라인단위로 읽고, 이렇게 읽은 문자열들을 저장하는 리스트를 sort 함수를 통해 정렬합니다.<br><br>
> 2. s1.txt 파일을 쓰기모드로 열어 정렬된 리스트를 writelines 함수를 이용하여 파일에 작성합니다. <br><br>
> 3. 그리고 s1.txt 파일에 내용이 잘 들어갔는지 확인하기 위해 읽기모드로 열어 read 함수를 통해 파일의 내용을 확인합니다.<br><br>

2) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.

> ham bird <br> 
> cat dog <br>
> pig ham <br>
> dog pig <br>

````python
li = []
with open('s.txt','r') as f:
    for tmp in f:
        li.append(tmp.replace('\n', ''))
    li.sort(key = lambda x: x[1:])
    
with open('s2.txt','w') as f:
    f.write('\n'.join(li))
    
# s2.txt 파일에 내용이 잘 들어갔는 지 확인
with open('s2.txt', 'r') as f:
    print(f.read()) 
````

> 1. 먼저 빈 리스트를 정의합니다. <br><br>
> 2. s.txt 파일을 읽기 모드로 열어 해당 파일을 for~in 구문으로 돌면서 하나의 라인의 마지막 개행문자 ```\n```을 replace 함수를 통해 없애줍니다. <br><br>
> 3. for~in 구문이 끝나면 sort 함수를 호출하는데 각 라인의 두번째 단어를 기준으로 정렬하기 위해 key를 정의합니다. <br><br>
> 4. 여기서 key는 lambda 함수로 정의하였는데, 각 라인의 두번째 단어 기준으로 정렬하는 것이여서 x[1:]으로 설정하였습니다. <br><br>
> 5. 마지막은 s2.txt 파일에 내용이 잘 들어갔는지 확인하기 위해 읽기모드로 열어 read 함수를 통해 파일 내용을 확인합니다. <br><br>

3) 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다.

> pig ham cat <br>
> dog ham bird <br>
> dog pig <br>

````python
word_onelist = []
with open('s.txt', 'r') as f:
    for line in f:
        word_onelist.extend(line.replace('\n','').split())

with open('s3.txt','w') as f:
    for i in range(int(len(word_onelist)/3 + 1)):
        f.write(' '.join(word_onelist[3 * i : 3 * (i+1)]) + '\n')
        
# s3.txt 파일에 내용이 잘 들어갔는 지 확인
with open('s3.txt', 'r') as f:
    print(f.read()) 
````

> 1. 먼저 빈 리스트를 정의합니다.<br><br>
> 2. s.txt 파일을 읽기 모드로 열어 해당 파일을 for~in 구문으로 돌면서 각 라인의 개행문자 ```\n```을 없애고 split으로 공백문자를 없애 extend 합니다. 이렇게 하면 파일에 들어있는 순수 문자열들이 하나의 리스트에 원소로 저장할 수 있습니다.<br><br> 
> ex) <br> 
pig ham <br> 
cat dog <br> 
ham bird <br> 
dog pig <br>을<br><br>
['pig', 'ham', 'cat', 'dog', 'ham', 'bird', 'dog', 'pig'] 으로 변경<br><br>
> 3. s3.txt 파일을 쓰기 모드로 열어 해당 파일을 for~in 구문으로 도는데 len(word_onelist)/3 + 1 까지 반복하면서 리스트의 3개의 문자열을 join 함수를 통해 병합하고 가장 마지막에는 개행 문자 ```\n```을 삽입합니다. 여기서 3개의 문자열을 검색하기 위해 인덱스를 ```[3 * i : 3 * (i+1)]```로 주었습니다.

````python
import hashlib 
import os
fileName = 'access.txt'

# 메뉴를 출력하는 함수
def printMenu():
    print("\n")
    print("Welcome to Our Service\n")
    print("1.Sign Up")
    print("2.Sign In")
    print("3.Quit")

    print("\n")
    return

# id가 존재하는지 확인하기 위한 함수
def isExistID(uid):
    with open(fileName, 'r') as file:
        for line in file:
            fid = line.split(':')[0]
            fid = fid[1:len(fid)-1]
            if(fid == uid):
                return True
        return False

    
# 회원가입 함수
def regist(uid, pw, name, school):
    with open(fileName, 'a') as file:
        file.write("[" + uid + "]" + ":" + "[" + hashlib.sha224(pw.encode()).hexdigest() 
                   + "]" + ","+  "[" + name + "]" + ","+ "[" + school + "]" + "\n")
 
        
# 로그인 체크 함수, id, pw 일치해야함  
def loginCheck(uid,pw):
    with open(fileName, 'r') as file:
        for line in file:
            # split 함수를 통해 라인별 회원 정보 추출
            fid = line.split(':')[0]
            fid = fid[1:len(fid)-1]
            fpw = line.split(':')[1].split(',')[0]
            fpw = fpw[1:len(fpw)-1]
            fname = line.split(':')[1].split(',')[1]
            if fid == uid:
                if fpw == hashlib.sha224(pw.encode()).hexdigest():
                    return fname
        return ''

# 최초 실행시 파일이 없다면 생성
if not os.path.isfile(fileName):
    with open(fileName, 'w') as file:
        file.write('')

while True:
    printMenu();
    user = input('Select Menu : ')
    if user == '1':
        uid = input('ID : ')
        while(isExistID(uid)):
            print('Sorry, th|e entered ID is already used.\n')
            uid = input('ID : ')
        pw = input('Password : ')
        name = input('Name : ')
        school = input('School : ')
        regist(uid, pw, name, school)
    
    elif user == '2':
        uid = input('ID : ')
        while(not isExistID(uid)):
            print('Sorry, you are not a registered member.\n')
            uid = input('ID : ')
            
        pw = input('Password : ')
        name = loginCheck(uid, pw)
        while name == '':
            print('Sorry, the entered password is not correct.\n')
            pw = input('Password : ')
            name = loginCheck(uid, pw)
        print ("Hello" + name + '!\n')
    elif user == '3':
        print ("Bye~\n\n")
        break
    else :
        print ("Sorry, the entered menu doesn't exist.\n")
        continue
````

> 1. 회원관리에 필요한 각 부분을 함수로 나누어 작성하였습니다. 먼저 메뉴를 출력하는 함수를 정의합니다. 그리고 아이디가 존재하는지 체크하는 함수를 정의합니다. 아이디가 존재하는지 체크하는 함수는 먼저 회원정보를 저장하는 파일을 읽기 모드로 열어 각 라인을 돌면서 라인에 가장 첫번째에 저장된 아이디를 뽑아냅니다. 아이디를 뽑아내는 방법은 정보가 [id]: [password], [name], [school]와 같이 저장되어있으므로 콜론(:)의 왼쪽값이 항상 아이디 정보라는 것을 알 수 있습니다. 콜론(:)을 기준으로 split 하고 인덱스 0의 값을 뽑아내면 됩니다. 뽑아냈을 때 '[' 문자와 ']' 문자가 존재하므로 해당 문자를 슬라이싱하여 순수 텍스트만 걸러내줍니다. 걸러진 텍스트와 사용자에게 입력받은 아이디와 같으면 True를 리턴합니다.<br><br>

### 3번 문제 

range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.
- 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
> range(stop) <br>
> range(start, stop[, step]) <br>
- 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
- 3) frange 사용 예
    - 3-1) frange(0.5)
        - [0.0, 0.1, 0.2, 0.3, 0.4]
    - 3-2) frange(1.0, 2.0)
        - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    - 3-3) frange(2.2, 4.0, 0.5)
        - [2.2, 2.7, 3.2, 3.7]

````python
def frange(start, stop = -1, step = -1):
    li = []
    
    # 인자를 하나만 입력받은 경우
    if stop == -1 and step == -1:
        stop = start
        start = 0.0
        step = 0.1
    
    # 인자를 두개 입력받은 경우
    elif stop >=0 and step == -1:
        step = 0.1
    
    if start > stop:
        return []
    
    i = 0.0
    while(start + step * i < stop):
        li.append(start + round(step * i, 15))
        i += 1.0
        
    return li
    

print(frange(0.5))
print(frange(1.0, 2.0))
print(frange(2.2, 4.0, 0.5))
print(frange(4, 6, 0.125))
print(frange(3, 4.33, 0.112))
print(frange(4, 1, 0.4))
````

> 1. 먼저 frange 함수는 들어오는 인자의 갯수가 달라지므로 디폴트 매개변수로 정의하였습니다. 만약 인자로 하나만 입력받은 경우에는(stop과 step이 디폴트 값인 -1임) 현재 start로 입력받은 값을 stop에 대입하고, start은 기본값인  0.0으로, step은 기본값인 0.1으로 초기화해줍니다.<br><br>
> 2. 다음으로 인자를 두개 입력받은 경우에는(stop값은 0보다 크거나 같고 step은 디폴트 값인 -1임) step값을 기본값인 0.1으로 초기화해줍니다.<br><br>
> 3. 만약 인자로 들어온 start 값이 stop 값보다 크다면 반환하는 값이 없으므로 빈 리스트를 반환해줍니다.<br><br>
> 4. 그리고 while문을 돌면서 start에 step 값을 더해 이 값이 stop보다 작다면 리스트에 append 합니다. 여기서 중요한 것은 내장함수인 round 함수를 이용한 것인데, 이는 반올림을 해줍니다. 소수점 자리 15번째까지 반올림을 수행해줍니다. (여기서 15번째까지 수행해준 이유는 0.3의 경우, float 값은 정확하지 않아서 0.30000000000000004와 같이 출력되는데, 15번째자리가 0이 끝나는 위치이므로 여기서 반올림을 수행하면 0.3이 됩니다.)<br><br>
> 5. 마지막으로 frange 함수의 결과가 저장된 리스트를 반환해주면 됩니다.

### 6번 문제

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

````python
def myfact(n):
    if n == 0: 
        return 1
    elif n == 1: 
        return 1
    else: 
        return n * myfact(n - 1)

print(myfact(3))
print(myfact(5))
print(myfact(7))
print(myfact(10))
print(myfact(50))
print(myfact(100))
print(myfact(300))
print(myfact(500))
print(myfact(700))
print(myfact(970))
print(myfact(971))
#print(myfact(972)) # 에러 발생 (maximum recursion depth 값을 넘음)
````

> 수학에서 factorial의 정의에 따라 myfact 함수의 인자에 0이 오는 경우 0!은 1이므로 1을 반환합니다. 또한 1!도 1이므로 1을 반환하고 위의 두 조건에 만족하지 않는다면 함수의 인자로 들어온 값 n과 myfact(n-1)을 곱한 값을 반환합니다.  <br><br>

> 위의 문제 [옵션]에서 n을 늘려보라고 하였는데, 제가 작동해보았을 때는 jupyter notebook에서 n이 971까지는 정상적으로 작동하였으나 972 부터 최대 재귀 깊이를 넘기 때문에 에러 메시지가 발생하는 것을 볼 수 있었습니다. 

````python
# 위의 maximum recursion depth 값을 넘는 에러를 해결하기 위해 밑의 두문장 추가

import sys
sys.setrecursionlimit(5000) # 최대 재귀 깊이를 증가시키기 위한 문장

def myfact(n):
    if n < 0: 
        return 0
    elif n == 1: 
        return n
    else: 
        return n * myfact(n - 1)
    
print(myfact(972))
````

- 1) 다음 사이트에서 제시되는 영어 불용어 (Stop words)를 참고하여 이전 숙제에서 구성했던 단어 사전에서 영어 불용어들을 모두 제거하는 코드를 추가하시오.

````python
# 영어 불용어가 저장되어있는 텍스트 파일 확인
with open('stopwordlist.txt','r') as f:
    print(f.read())
````

 이 과제를 수행하기 위하여 과제에서 참조된 사이트 ```http://www.lextek.com/manuals/onix/stopwords1.html ``` 을 기준으로 영어 불용어를 텍스트 파일에 저장하였습니다.

````python
def getStopWord(): # 영어 불용어를 반환하기 위한 함수
    with open('stopwordlist.txt','r') as f: # 영어 불용어가 저장되어있는 텍스트 파일을 읽기 모드로 엶
        stopword = f.read().split()
    return stopword # 영어 불용어 리스트 반환
````

> - getStopWord() 함수
    - 영어 불용어가 저장되어있는 텍스트 파일을 읽기 모드로 열어 split 함수를 이용하여 공백문자를 제거하였고, 이를 리스트로 반환하였습니다.

- 2) 각 URL로 지정된 웹페이지의 HTML 소스를 파일로 저장하시오.

    - URL이 http://URL 이라면 파일명은 URL.html 이다.
    - 예: URL이 http://www.cnn.com 이라면 파일명은 www.cnn.com.html 이다.

````python
from urllib import request 

def getUrlname(url): # URL을 변경하기 위한 함수
    # http:// 나 https://을 제거하기 위해 //을 기준으로 기존 url을 슬라이싱
    # 디렉토리 구분을 나타내는 문자 '/'은 파일명에 들어갈 수 없으므로 언더바('_')로 변경
    urlname = url[url.index('//') + 2 :]
    urlname = urlname.replace('/','_')
    return urlname

def saveSource(urlname, source):
    # source 저장
    with open(urlname + '.html', 'w', encoding = 'utf8') as f:
        f.write(source)
````

> - getUrlname() 함수
    - 해당 함수를 작성하기 위해 url을 받아와 index 함수를 이용하여 ```http://```나 ```https://```을 제거하기 위하여 '//' 문자를 기준으로 해당 문자가 시작하는 위치를 알아내고 해당 위치에서 2를 더해 ```http://```나 ```https://```을 제거한 urlname을 반환하였습니다. 또한 디렉토리 구분을 나타내는 문자 '/'는 파일명에 들어갈 수 없으므로 언더바 '_'로 변경하였습니다.<br><br>

> - saveSource() 함수
    - getUrlname 함수에서 반환한 urlname 뒤에 '.html'을 더한 값을 파일 이름으로 지정해주고 해당 파일을 쓰기 모드로 열어 HTML 소스를 작성해줍니다.

- 3) 단어의 출현빈도가 담긴 사전 객체를 위 HTML 소스 파일과 동일한 폴더에 파일로 저장하시오.
    - 파일입출력 (E-learning 13주차) 마지막에 학습한 pickle 모듈을 활용하시오.
    - URL이 http://URL 이라면 사전 객체를 담고 있는 파일명은 URL.words.frequency 이다.
    - 예: URL이 http://www.cnn.com 이라면 파일명은 www.cnn.com.words.frequency 이다.`
            

````python
import pickle

def saveWordFreq(urlname, wordList):
     # 단어 빈도수를 파일에 저장
    with open(urlname + '.words.frequency', 'wb') as f:
        pickle.dump(wordList, f)
````

> - saveWordFreq 함수
    - 단어의 출현빈도가 담긴 사전 객체를 urlname 뒤에 '.words.frequency'를 더한 값을 파일 이름으로 지정해주고 해당 파일을 쓰기 모드로 열어 사전 객체를 파일에 작성해줍니다.

````python
# 이전 Assignment 3의 Incremental Project 소스코드 일부
        
def deleteRemark(source):
    # 주석 제거
    while ('<!--' in source):
        # 주석은 <!-- 과 --> 의 쌍으로 나타내어지므로 index 함수를 사용해 주석을 찾아 없앤다
        left = source.index('<!--')
        right = source.index('-->')
        source = source[:left] + source[right+3:]
    return source
    
def deleteHtmlEntities(source):
    # html &nbsp; &quot; &lt; &gt; &amp; 제거
    htmlEntities = ['&nbsp;','&quot;','&lt;','&gt;','&amp;']
    for entity in htmlEntities:
        while(entity in source):
            left = source.index(entity)
            right = source.index(entity)
            source = source[:left] + source[right+len(entity):]    
    return source

def deleteHtmlTag(source, text):
     # <script> 태그가 없어질때까지 반복 (javascript 태그)
    while ('<script' in source):
        # <script> 태그 뒤에는 문장이 올 수 있으므로 <script 으로 검사해준다 
        # <script> ~ </script> 사이에 오는 문장은 순수 텍스트가 아니므로 무시한다
        left = source.index('<script')
        right = source.index('</script>')
        source = source[:left] + source[right+9:]
    
    # <style> 태그가 없어질때까지 반복 (css 태그)
    while ('<style' in source):
        # <style> 태그 뒤에는 문장이 올 수 있으므로 <style 으로 검사해준다 
        # <style> ~ </style> 사이에 오는 문장은 순수 텍스트가 아니므로 무시한다
        left = source.index('<style')
        right = source.index('</style>')
        source = source[:left] + source[right+8:] 
    
    # 태그는 < 과 > 의 쌍으로 나타내어지므로 index 함수를 사용해 태그 부분을 찾아 없앤다.
    while('<' in source):
        left = source.index('<')
        right = source.index('>')
        text = text+' '+source[:left]
        source = source[right+1:]
    return source, text
````

> 이전 Assignment 3의 Increamental Project 소스코드를 가져온 것입니다. 이는 소스코드의 주석, HTML ENTITIES, HTML TAG를 지우기 위한 함수입니다. 이전 과제에서 설명한 부분이므로 설명을 생략하겠습니다.

````python
import string  

def IncrementalProject(url):
    stopword = getStopWord()
    
    f = request.urlopen(url) 
    source = f.read().decode('utf-8')
    text = ''
    
    urlname = getUrlname(url)
    
    # source를 파일에 저장
    saveSource(urlname, source)
 
    # 주석 제거
    source = deleteRemark(source)
    
    # HtmlEntities 제거
    source = deleteHtmlEntities(source)
    
    # HtmlTag 제거
    source, text = deleteHtmlTag(source, text)
    
    wordList = {} # { 단어(키) : 단어의 출현 빈도(벨류) } 형식의 사전 정의
    puncList = list(string.punctuation)
    # 구두문자와 html entity 추가
    puncList.extend(['’','‘','···','【','】','…','·','»','¢','£','¥','€','®','ⓒ','©']) 

    for t in text.split(): # 현재 text에 들어 있는 문장은 순수 텍스트이므로 공백을 기준으로 자름
        for char in puncList: 
            t = t.replace(char, '')
        t = t.lower() # 모든 단어를 대소문자를 구분하지 않고 단어를 저장함
        if len(t) > 0 and not t in stopword:
            if (not t in wordList.keys()):
                wordList[t] = 1
            else:
                wordList[t] = wordList[t] + 1
   
    print(wordList)
    
    # 단어 빈도수를 저장하는 사전 객체를 파일에 저장
    saveWordFreq(urlname, wordList)
````


> ```  f = request.urlopen(url) 
> source = f.read().decode('utf-8') ```

> 3. 주석을 제거하기 위한 deleteRemark 함수와 HtmlEntities을 제거하기 위한 deleteHtmlEntities, HtmlTag을 제거하기 위한 deleteHtmlTag을 각각 호출하여 해당 페이지의 소스코드를 수정해줍니다.<br><br>

> 4. 단어와 단어 출현빈도를 저장하기 위한 사전 wordList을 정의하고, string 모듈의 punctuation을 통해 구두문자를 리스트 puncList으로 반환해줍니다. 해당 리스트에 '’','‘','···','【','】','…','·','»','¢','£','¥','€','®','ⓒ','©'와 같은 문자를 추가하여줍니다.<br><br>

- 4) 최소 5개 이상의 영어 웹 사이트 각각에 대한 HTML 소스 파일과 단어 출현빈도 파일을 저장하시오.
    - 즉, 총 10개의 파일을 동일한 폴더에 생성하시오.
    - [주의] 영어 웹사이트 URL로만 5개 이상
        - 가능하면 주제가 서로 다른 웹 사이트들로 선정<br><br>
- 5) 위 문제에서 저장한 모든 words.frequency 파일들을 객체로 다시 로드하여 본인이 저장하여 분석한 사이트들 각각에 대해서 가장 많이 출현한 단어 3개씩를 뽑아 제시하시오.
    - 반드시 pickle 모듈로 저장한 5개 이상의 words.frequency를 다시 5개 이상의 사전 객체로 로드 하는 코드가 추가되어야 함


````python
# 사전에 들어있는 단어의 최빈도 상위 x개를 구하는 함수
def getTopWord(dic, x):
    lists = list(dic.keys())
    maxValue = 0
    maxKey = ''
    for i in range(len(lists)): 
        if dic[lists[i]] > maxValue: 
            # 내림 차순으로 출력하기 위함
            # 처음에 출현빈도를 기준으로 가장 출현빈도가 높은 값을 먼저 출력하고 아래에서 재검색을 막기 위해 해당 단어의 출현 빈도를 -1으로 설정
            # 다음으로 출현빈도를 기준으로 높은 값을 출력 ... 반복
            # 이렇게 하면 내림 차순으로 출력 가능
            maxValue = dic[lists[i]]
            maxKey = lists[i]
    print (maxKey , maxValue)
    
    # 재검색을 막기 위해 -1 으로 설정해놈 (이미 최빈도 단어로 선택된 단어임을 나타냄) 
    dic[maxKey] = -1
    if(x > 1):
        getTopWord(dic, x - 1)
    
    # 복구시킴
    dic[maxKey] = maxValue 
````

> 1. 사전에 들어있는 단어의 최빈도 상위 x개를 구하는 함수 getTopWord을 정의합니다. 먼저 해당 함수에서는 단어의 출현빈도를 저장하는 사전 객체의 단어들만 리스트로 저장합니다. 그 리스트의 길이만큼 반복문들 돌면서 현재 저장되어있는 출현빈도가 높은 단어보다 더 출현빈도가 높은 단어가 있으면 해당 벨류를 maxValue에 저장하고, 해당 키도 mayKey에 저장하고, 이를 출력합니다. <br><br>

> 2. 현재 출현빈도가 가장 높은 단어의 재검색을 막기 위하여 -1으로 설정해놉니다. 이렇게 하는 이유는 이미 최빈도 단어로 선택되어 출력되었기 때문에 재검색을 맏기 위함입니다. <br><br>

> 3. 그리고 x가 1보다 크다면 해당 함수를 x를 하나 감소하여 재귀적으로 호출해줍니다. 그리고 위에서 단어의 재검색을 막기 위해 출현빈도 값을 -1으로 설정한 것을 다시 복구 시킵니다.

- 6) 간단한 검색엔진 코딩 (Like Google!!!)
    - 사용자에게 임의의 단어 1개를 입력받으시오.
    - 입력 받은 단어와 유사도(Similarity)가 높은 웹 사이트 기준으로 위 4)에서 미리 지정해 놓은 5개 이상의 웹 사이트 URL들을 일렬로 나열하여 출력하시오.
        - 단어와 웹 사이트 간의 유사도는 본인이 스스로 정하시오.
            - 유사도를 정하는 최소한의 기준은 단어 출현 빈도를 기반으로 해야 하며, 이외의 본인이 생각하는 방안이 있으면 함께 사용해도 됨.
            - 반드시 코딩 내에 similarity 라는 사전 자료 구조가 있어야 함 (즉, 다음과 유사한 코드가 있어야 함).
                - similarity = {}
                - ...
                - similarity[url1] = x1
                - ...
                - similarity[url2] = x2
                - ...
        - 위 코드에서 x1 과 x2는 입력단어와 해당 URL의 웹 사이트 간의 similarity를 나타내는 수치값
    - 유사도가 높은 웹 사이트가 상위에 출력되어야 함 (즉, 유사도 기준 내림 차순)
        - 유사도가 동일한 웹 사이트들에 대해서는 임의 배치함.

````python
# 간단한 검색엔진 코딩
word = input('word : ').lower() # 입력받은 키워드는 대소문자 구별하지 않음
print ("\n...검색 중 입니다...\n")
similarity = {}
tmp = []
for i in range(len(urlList)):
    similarity[urlList[i]] = 10 # 기본 유사도를 10으로 줌
    tmp = list(urlWords[i].keys())
    print ("\nurl : " + urlList[i])
    print ("\n해당 키워드가 포함된 단어 항목입니다\n")
    for j in range(len(tmp)):
        if tmp[j].find(word) != -1 and tmp[j] != word:
            print(tmp[j])
            similarity[urlList[i]] += urlWords[i][tmp[j]] 
            # 입력받은 키워드가 단어 일부일 경우, 해당 단어의 출현 빈도를 더해줌
        
       
        elif tmp[j] == word:
            print(tmp[j])
            similarity[urlList[i]] += (urlWords[i][tmp[j]] + 20) 
            # 입력받은 키워드가 단어와 같을 경우에는, 해당 단어의 출현 빈도에 보너스 점수 20을 더한 것을 더해줌
    print('\n***********************************************\n') 
````

> 1. 간단한 검색엔진을 코딩하기 위하여 사용자로부터 키워드를 입력하게 합니다. <br><br>

> 2. 그리고 유사도를 저장할 similarity 사전 객체를 정의하여 줍니다. <br><br> 

> 3. 위에서 저장한 웹 사이트의 리스트 urlList의 갯수만큼 돌면서 해당 url의 similarity 값을 기본 유사도 점수인 10으로 줍니다. 그리고 해당 url의 출현빈도를 저장하는 사전 객체의 키 값을 리스트로 저장합니다. 해당 리스트의 길이만큼 돌면서 리스트의 키 값에 사용자가 입력한 키워드 값이 포함되어있다면(리스트 키 값과 키워드 값이 동일한 것은 아님) 해당 단어의 출현 빈도를 similarity 값에 저장해줍니다.<br><br> 

> 4. 만약 리스트 키 값과 키워드 값이 동일하다면 해당 단어의 출현 빈도에 20(리 값과 키워드 값이 동일하기 때문에 유사도를 조금 더 높게 설정해주기 위하여 20점으로 설정하였습니다)을 더한 값을 similarity 값에 저장해줍니다.
