### 5번 문제

- 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
    - 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
               
                >>> addall([1])
                1
                
                >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
                45        
    - 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
       
                >>> addallodd([1])
                1
                >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
                25
    - [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
    - [참고 2]: sum() 등의 내장함수 사용 가능함

````python
# 1번 for ~ in 구문
def addall(li):
    sum = 0;
    for i in range(len(li)):
        sum += li[i]
    return sum
print (addall([1]))
print (addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

````python
# 1번 리스트 내포
def addall(li):
    return sum([i for i in li])
print (addall([1]))
print (addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

> - 5번 문제의 1) 의 경우, for~in을 사용할 때는 먼저 정수의 합을 담고 있을 변수 sum을 0으로 초기화해주고, for~in 구문을 리스트의 원소 갯수만큼 돌면서 리스트의 원소들을 sum에 더해주었습니다. 마지막으로 원소들의 합을 담고 있는 sum 변수를 반환하는 것으로 addall 함수를 작성하였습니다. <br><br>
> - 반면에 리스트 내포 방식을 사용할 때에는 리스트 내의 원소들의 합을 구하기 위해 sum이라는 내장함수를 이용하였습니다. 리스트를 인자로 받아온 후, 리스트 내의 숫자들로 리스트를 생성하여 그 리스트를 sum이라는 내장함수의 인자로 줍니다. 그리고 그 결과를 반환하는 것으로 addall 함수를 작성하였습니다.

````python
# 2번 for ~ in 구문
def addallodd(li):
    sum = 0
    for i in range (len(li)):
        if(li[i] % 2 == 1):
            sum += li[i]
    return sum
print (addallodd([1]))
print (addallodd([1,2,3,4,5,6,7,8,9]))
````

````python
# 2번 리스트 내포
def addallodd(li):
    return sum([i for i in li if(i % 2 == 1)])
print (addallodd([1]))
print (addallodd([1,2,3,4,5,6,7,8,9]))
````

> - 5번 문제의 2) 의 경우, for~in을 사용할 때에는 먼저 홀수의 합을 담고 있을 변수 sum을 0으로 초기화해주고, for~in 구문을 리스트의 원소 갯수만큼 돌면서, 만약 리스트의 원소가 2로 나눈 나머지가 1(홀수가 되기 위한 조건)라면, sum에 해당 리스트의 원소를 더해줍니다. 그리고 마지막으로 홀수의 합을 담고 있는 sum 변수를 반환하는 것으로 addallodd 함수를 작성하였습니다. <br><br>
> - 반면에 리스트 내포 방식을 사용할 때에는 리스트 내의 홀수의 합을 구하기 위해 sum이라는 내장함수를 아용하였습니다. 리스트를 인자로 받아온 후, 리스트 내의 숫자들로 리스트를 생성하는데, 이 과정에서 해당 리스트 원소를 2로 나눈 나머지가 1(홀수가 되기 위한 조건)인 경우에만 리스트의 원소로 들어갈 수 있도록 하였고, 이렇게 만들어진 리스트를 sum이라는 내장함수의 인자로 주고, 그 결과를 반환하는 것으로 함수를 작성하였습니다.

     - 영어 대소문자와 띄어쓰기만으로 이루어진 문장이 주어진다. 이 문장에는 몇 개의 단어가 있을까? 이를 구하는 프로그램을 작성하시오. 단, 단어는 띄어쓰기 하나로 구분된다고 생각한다.
     
        - [입력] 첫 줄에 영어 대소문자와 띄어쓰기로 이루어진 문장이 주어진다. 이 문장의 길이는 1,000,000을 넘지 않는다.
        - [출력] 첫째 줄에 단어의 개수를 출력한다.

````python
def countVoca(item):
    sp = item.split()
    cnt = len(sp)
    return cnt

sentence = input()
print(countVoca(sentence))
````

> - 먼저 문장의 단어 개수를 반환해주는 countVoca 함수를 정의합니다. countVoca 함수에서는 split 함수를 이용하여 공백(띄어쓰기)를 기준으로 문장을 잘라줍니다. 이렇게 하게 되면 문장 내에 존재하는 단어들이 리스트 형태로 반환되고, 만들어진 리스트의 원소 개수를 len 함수를 이용하여 구해주면 해당 문장의 단어의 개수가 몇 개인지 알게 됩니다. <br><br>
> - countVoca 함수 외부에서는 사용자에게 영어 대소문자와 띄어쓰기만으로 이루어진 문장을 입력받기 위해 input 함수를 통하여 입력받고. 입력받은 문장을 countVoca 함수의 인자로 주어지게 되면, 결과적으로 사용자가 입력한 문장의 단어 개수를 구할 수 있게 됩니다.

### 9번 문제

- [Incremental Project] 이전 Assignment 2의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그 및 CSS와 Javascript를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하고 단어들의 총 개수를 출력하는 프로그램을 작성하는 것이었다. 이번에는 그 마지막 숙제를 그대로 확장하여 웹 URL로 지정된 웹페이지 내 순수 텍스트 안에 존재하는 각 단어들에 대해 다음 요구사항 대로 출력하는 프로그램을 작성하시오.
    - 요구사항 1. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.
        - 예: ['world!', ':', '+hello+', '~python$$$', '=', 'world'] ---> ['world', 'hello', 'python', 'world']
        - 모든 punctuation을 확인하는 방법은 아래와 같다.
                  import string
                  print(string.punctuation)
    - 요구사항 2. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.
        - 예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '지향']
   - 요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.
       - 예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}
    - 요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.
        - http://www.naver.com
        - http://www.daum.net
        - http://www.nytimes.com

````python
from urllib import request 
import string

def IncrementalProject(url):
    f = request.urlopen(url)
    source_bytes = f.read()
    source = source_bytes.decode('utf-8')
    text = ''

    # 주석 제거
    while ('<!--' in source):
        # 주석은 <!-- 과 --> 의 쌍으로 나타내어지므로 index 함수를 사용해 주석을 찾아 없앤다
        left = source.index('<!--')
        right = source.index('-->')
        source = source[:left] + source[right+3:]
    #print(source)
    
    # <script> 태그가 없어질때까지 반복 (javascript 태그)
    while ('<script' in source):
        # <script> 태그 뒤에는 문장이 올 수 있으므로 <script 으로 검사해준다 
        # <script> ~ </script> 사이에 오는 문장은 순수 텍스트가 아니므로 무시한다
        left = source.index('<script')
        right = source.index('</script>')
        source = source[:left] + source[right+9:]
    
    # <style> 태그가 없어질때까지 반복 (css 태그)
    while ('<style' in source):
        # <style> 태그 뒤에는 문장이 올 수 있으므로 <style 으로 검사해준다 
        # <style> ~ </style> 사이에 오는 문장은 순수 텍스트가 아니므로 무시한다
        left = source.index('<style')
        right = source.index('</style>')
        source = source[:left] + source[right+8:] 
 
    
    # 태그는 < 과 > 의 쌍으로 나타내어지므로 index 함수를 사용해 태그 부분을 찾아 없앤다.
    while('<' in source):
        left = source.index('<')
        right = source.index('>')
        text = text+' '+source[:left]
        source = source[right+1:]
      

    wordList = {} # { 단어(키) : 단어의 출현 빈도(벨류) } 형식의 사전 정의
    puncList = list(string.punctuation)
    puncList.extend(['’','‘','···','【','】','…','·']) # 구두문자 추가

    for t in text.split(): # 현재 text에 들어 있는 문장은 순수 텍스트이므로 공백을 기준으로 자름
        for char in puncList: 
            t = t.replace(char, '')
        
        if (len(t) > 0):
            if (not t in wordList.keys()):
                wordList[t] = 1
            else:
                wordList[t] = wordList[t] + 1
    
    print(wordList)          
````

    해당 문자열 ```<!--```이 감지된다면 감지된 인덱스를 left 변수에 저장하고, 
    문자열 ```-->```이 감지된다면 감지된 인덱스를 right 변수에 저장합니다.

    순수 텍스트를 답는 text 변수에 < 문자가 감지되기 이전까지의 소스코드를 text에 덧붙여줍니다. 
    그리고 soruce는 > 문자가 감지된 이후부터로 갱신해주면 됩니다.<br><br>

    만약 replace된 단어의 길이가 0보다 크고 순수 텍스트와 순수 텍스트 출현 빈도를 저장하는 사전 wordList에 들어가 있지 않으면 
    해당 단어를 키 값으로 하는 벨류 값을 1로(처음 들어가는 것이므로) 줍니다.  반면에 replace된 단어의 길이가 0보다 크지만, 사전 wordList에 들어가 있으면 해당 단어를 키 값으로 하는 벨류 값을 1 증가시켜줍니다.
